```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x),"-")
plt.plot(x, np.cos(x),"--")

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(("sin","cos"))


```

```python
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))
plt.xlabel("x axis")

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.xlabel("y axis")

plt.legend(("sin","cos"))
```

```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
# for panda dataframe
iris["petal_length"].plot()
```

```python
# this one is better.
sns.lineplot(data=iris, y="petal_length", x=iris.index, hue="species")
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(iris["sepal_length"], iris["sepal_width"])
ax.set_title("iris dataset")
ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
```

```python
 iris.corr()
```

```python
sns.heatmap(iris.corr(), annot=True)
```

```python
sns.boxplot(data=iris, y="sepal_length", x="species")
```
